```python
#Ejercicio 1.- Ingreso a una boveda.

#Pedro y Jes√∫s son dos hermanos, hijos de Manuel, y quienes tienen acceso a una b√≥veda en un exclusivo banco.
#El due√±o original de esta b√≥veda es su padre, y en su testamento ha especificado las siguientes condiciones para que el banco
#le permita acceder a sus hijos al contenido de dicha b√≥veda. Las condiciones son las siguientes:

#a. Juan debe ser mayor de 35 a√±os, o Pedro debe ser menor de 50 a√±os.

#b. La suma de sus edades debe ser un n√∫mero par.

#c. Manuel debe tener m√≠nimo 1 nieto.

#Construya una funci√≥n que lea la edad de Juan, la edad de Pedro, el n√∫mero de nietos y calcule, usando condicionales, si tienen acceso (SI) o no (NO) a la b√≥veda

def b√≥veda():
  while True:
    try:
      edad_juan = int(input("Si la edad de Juan es: "))
      edad_pedro = int(input("La de Pedro es: "))
      nietos = int(input("Y la cantidad que ambos tienen de hijos es: "))
      break
    except:
      print("Tiene que ser un n√∫mero entero, intenta otra vez")

  if (edad_juan > 35 or edad_pedro < 50) and (edad_juan + edad_pedro) % 2 == 0 and nietos >= 1:
    return "SI"
  else:
    return "NO"

print('¬øTienen Pedro y Jes√∫s acceso a la b√≥veda?', b√≥veda())
```

```python
# Ejercicio 5.- MCD y MCM.

# Dados dos n√∫meros a y b, determine:

# El m√°ximo com√∫n divisor.
# El m√≠nimo com√∫n m√∫ltiplo.
# Almacene los resultados en un diccionario e imprima en pantalla el contenido del diccionario

def calcular_mcd_mcm():
  a=int(input("Ingrese el primer n√∫mero: "))
  b=int(input("Ingrese el segundo n√∫mero: "))
  def mcd(a,b):
    while b:
      a, b = b, a % b
    return a
  maximo_comun_divisor = mcd(a, b)

  minimo_comun_multiplo = (a * b) // maximo_comun_divisor

  resultados = {
        "MCD": maximo_comun_divisor,
        "MCM": minimo_comun_multiplo
    }



  return resultados
print(calcular_mcd_mcm())
```

```python
# Ejercicio 6.- Filtrado de datos.

# Escriba una funci√≥n que reciba una lista de datos (la de abajo) y devuelva listas con los datos filtrados. Este programa debe separar los datos en int, str, bool, float.

# Hint: use la funci√≥n append para insertar datos en las listas
def filtrar_datos(datos):

    enteros = []
    cadenas = []
    booleanos = []
    flotantes = []

    for dato in datos:
        if type(dato) == int:
            enteros.append(dato)
        elif type(dato) == str:
            cadenas.append(dato)
        elif type(dato) == bool:
            booleanos.append(dato)
        elif type(dato) == float:
            flotantes.append(dato)

    resultados = {
        "int": enteros,
        "float": flotantes,
        "str": cadenas,
        "bool": booleanos
    }

    return resultados

datos = [25, "Matem√°ticas", 3.141592, True, 8, False, "Nombres", 2.718, "UNAM", 0,
         9.81, "CdMx", 7e12, 12.5, "2025", True, "poblaci√≥n", 0.01, 42, False]

datos_filtrados = filtrar_datos(datos)
print(datos_filtrados)
```

```python
# Ejercicio 7.- C√°lculo de la distancia entre dos puntos mediante la f√≥rmula de Haversine.

# La f√≥rmula de Haversine permite calcular la distancia entre dos puntos sobre la superficie de la Tierra, a partir de sus coordenadas geogr√°ficas.
#La f√≥rmula est√° dada por la siguiente expresi√≥n:
# d=2Rarcsin(haversin(œÜ2‚àíœÜ1)+cos(œÜ1)cos(œÜ2)haversin(Œª2‚àíŒª1)‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àí‚àö)

# donde:
# d = distancia entre dos puntos (en km),
# R = radio de la Tierra (ùëÖ‚âà6378.1),
# ùúë1, ùúë2 = latitudes de los dos puntos.
# ùúÜ1, ùúÜ2 = longitudes de los dos puntos (en radianes).

# La funci√≥n de Haversine se define como:
# haversin(Œ±)=sin2(Œ±2)

# Para convertir de grados a radianes se utiliza la relaci√≥n:
# radianes=grados√óœÄ180

# Construya una funci√≥n que permita calcular la distancia entre dos pares de coordenadas. Utiliza la librer√≠a math para las funciones trigonom√©tricas y de ra√≠z cuadrada.
# Ejemplo de prueba. Coordenadas de Par√≠s y Roma por latitud y longitud: Par√≠s (48,9, 2,4) y Roma (41,9, 12,5). Resultado: distancia=1107 km.
# Proponga 3 pares de coordenadas. Al menos uno debe ser respecto a la Ciudad de M√©xico ( Latitud=19.43, Longitud=-99.13 )
import math

def haversine(lat1, lon1, lat2, lon2):

    R = 6378.1  # Radio de la Tierra en km

    # Convertir grados a radianes
    lat1_rad = math.radians(lat1)
    lon1_rad = math.radians(lon1)
    lat2_rad = math.radians(lat2)
    lon2_rad = math.radians(lon2)

    # Diferencias de latitud y longitud
    delta_lat = lat2_rad - lat1_rad
    delta_lon = lon2_rad - lon1_rad

    # Funci√≥n haversin
    haversin_lat = math.sin(delta_lat / 2)**2
    haversin_lon = math.sin(delta_lon / 2)**2

    # F√≥rmula de Haversine
    a = haversin_lat + math.cos(lat1_rad) * math.cos(lat2_rad) * haversin_lon
    distancia = 2 * R * math.asin(math.sqrt(a))

    return distancia

# Ejemplo de prueba con Par√≠s y Roma
paris_lat = 48.9
paris_lon = 2.4
roma_lat = 41.9
roma_lon = 12.5

distancia_paris_roma = haversine(paris_lat, paris_lon, roma_lat, roma_lon)
print(f"Distancia entre Par√≠s y Roma: {distancia_paris_roma:.0f} km")

# Coordenadas propuestas
# 1. Ciudad de M√©xico a Nueva York
mexico_lat = 19.43
mexico_lon = -99.13
nueva_york_lat = 40.71
nueva_york_lon = -74.01

distancia_mexico_ny = haversine(mexico_lat, mexico_lon, nueva_york_lat, nueva_york_lon)
print(f"Distancia entre Ciudad de M√©xico y Nueva York: {distancia_mexico_ny:.0f} km")

# 2. Nueva York a Madrid
madrid_lat = 40.42
madrid_lon = -3.70

distancia_ny_madrid = haversine(nueva_york_lat, nueva_york_lon, madrid_lat, madrid_lon)
print(f"Distancia entre Nueva York y Madrid: {distancia_ny_madrid:.0f} km")

# 3. Madrid a Buenos Aires
buenos_aires_lat = -34.60
buenos_aires_lon = -58.38

distancia_madrid_buenos_aires = haversine(madrid_lat, madrid_lon, buenos_aires_lat, buenos_aires_lon)
print(f"Distancia entre Madrid y Buenos Aires: {distancia_madrid_buenos_aires:.0f} km")
```

```python
# Ejercicio 8.- La serie de Fibonacci.

# Realice un programa que calcule la serie de Fibonacci y la imprima en pantalla, tenga en cuenta las restricciones de la variable de entrada.
def fibonacci():
    while True:
        try:
            n = int(input("Ingrese el n√∫mero de t√©rminos de la serie de Fibonacci (entero no negativo): "))
            if n < 0:
                print("Error: El n√∫mero de t√©rminos debe ser no negativo.")
                continue
            break
        except ValueError:
            print("Error: Ingrese un n√∫mero entero v√°lido.")

    a, b = 0, 1
    serie = []

    for _ in range(n):
        serie.append(a)
        a, b = b, a + b

    print(serie)
fibonacci()
```

```python
# Ejercicio 9.- Mini-cat√°logo de sismos.

# En este ejercicio se muestran 4 listas que representan un conjunto de sismos con la siguiente informaci√≥n (fecha, zona, magnitud y profundidad).
#Hacer un programa que haga lo siguiente:
# 1.- Convierte la lista de magnitues de str a float.
# 2.- Los nombres de los estados deben iniciar con may√∫scula (utilice una funci√≥n lambda).
# 3.- Crea una lista de llamadas intensidad (con la misma longitud de la lista de magnitudes) y coloca etiquetas de acuerdo al siguiente criterio:
  # Magnitud < 3.5 "Muy Leve"
  # 3.5 <= Magnitud < 4.5 "Nivel"
  # 4.5 <= Magnitud < 3.5 "Moderado"
  # Magnitud > 5.5 "Fuerte"
# 4.- Con un ciclo For (y condicionales si es necesario), genera un diccionario "conteo_por_zona" con el n√∫mero de eventos por zona.
# 5.- Ordena los eventos por magnitud descendente y muestra los 3 sismos m√°s grandes con el siguiente formato:
  # AAAA-MM-DD | Zona (Estado) | Magnitud | Profundidad (km).
# Sugerencia: Recuerde las funciones predefinidas de python map, filter, len, entre otras

Fechas = ["2025-03-01","2025-03-02","2025-03-03","2025-03-04","2025-03-05",
    "2025-03-06","2025-03-07","2025-03-08","2025-03-09","2025-03-10"]

Zonas = ["guerrero","oaxaca","chiapas","michoac√°n","guerrero",
    "oaxaca","chiapas","guerrero","oaxaca","chiapas"]

magnitudes = ["3.8","4.2","5.0","2.9","4.8","3.5","4.1","5.3","4.0","3.0"]

prof_km = [15, 80, 12, 5, 35, 120, 60, 25, 10, 90] #Profundidad en kil√≥metros.

# 1. Convierte la lista de magnitudes de str a float.
magnitudes = list(map(float, magnitudes))

# 2. Los nombres de los estados deben iniciar con may√∫scula (utiliza una funci√≥n lambda).
Zonas = list(map(lambda zona: zona.capitalize(), Zonas))

# 3. Crea una lista de llamadas intensidad (con la misma longitud de la lista de magnitudes) y coloca etiquetas de acuerdo al siguiente criterio:
intensidades = []
for i in range(len(magnitudes)):
     magnitud = magnitudes[i]
     zona = Zonas[i]
     if magnitud < 3.5:
         intensidad = "Muy Leve"
     elif 3.5 <= magnitud < 4.5:
         intensidad = "Leve"
     elif 4.5 <= magnitud <= 5.5:
         intensidad = "Moderado"
     else:
         intensidad = "Fuerte"
     intensidades.append(f"{zona}: {intensidad}")

# 4. Con un ciclo For (y condicionales si es necesario), genera un diccionario "conteo_por_zona" con el n√∫mero de eventos por zona.
conteo_por_zona = {}
for zona in Zonas:
    if zona in conteo_por_zona:
        conteo_por_zona[zona] += 1
    else:
        conteo_por_zona[zona] = 1

# 5. Ordena los eventos por magnitud descendente y muestra los 3 sismos m√°s grandes con el siguiente formato: AAAA-MM-DD | Zona (Estado) | Magnitud | Profundidad (km).
sismos = list(zip(Fechas, Zonas, magnitudes, prof_km))
sismos_ordenados = sorted(sismos, key=lambda sismo: sismo[2], reverse=True)

print("\nLos 3 sismos m√°s grandes:")
for i in range(min(3, len(sismos_ordenados))):
    fecha, zona, magnitud, profundidad = sismos_ordenados[i]
    print(f"{fecha} | {zona} | {magnitud} | {profundidad} km")

# Imprimir las listas transformadas (opcional, si quieres verificar)
print("\nLista de magnitudes (float):", magnitudes)
print("Lista de zonas (capitalizadas):", Zonas)
print("Lista de intensidades:", intensidades)
print("Conteo por zona:", conteo_por_zona)
```

```python
# Ejercicio 10.- Operaciones vectoriales.

# Dadas las siguientes listas:
# a = [2, -1, 4]
# b = [4, -2, 8]
# c = [-5, 2, 1]

# Calcule el producto escalar:
# ab
# bc
# ca

# Calcular el productor vectorial:
# a x b
# b x c
# a x c

# Utilice las listas de inicio para probar su c√≥digo. Posteriormente, genere una funci√≥n que reciba como entrada las listas (vectores) y haga las operaciones vectoriales.
# Indique si algunos de los vectores son paralelos o perpendiculares
import numpy as np

def operaciones_vectoriales(a, b, c):

    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    # Producto escalar
    ab = np.dot(a, b)
    ac = np.dot(a, c)
    bc = np.dot(b, c)

    # Producto vectorial
    axb = np.cross(a, b)
    bxc = np.cross(b, c)
    axc = np.cross(a, c)

    # An√°lisis de paralelismo y perpendicularidad
    def son_paralelos(v1, v2):
        if np.allclose(v1 / np.linalg.norm(v1), v2 / np.linalg.norm(v2)) or \
           np.allclose(v1 / np.linalg.norm(v1), -v2 / np.linalg.norm(v2)):
            return True
        return False

    def son_perpendiculares(v1, v2):
        return np.isclose(np.dot(v1, v2), 0)

    paralelos = []
    perpendiculares = []

    if son_paralelos(a, b):
        paralelos.append("a y b")
    if son_paralelos(a, c):
        paralelos.append("a y c")
    if son_paralelos(b, c):
        paralelos.append("b y c")

    if son_perpendiculares(a, b):
        perpendiculares.append("a y b")
    if son_perpendiculares(a, c):
        perpendiculares.append("a y c")
    if son_perpendiculares(b, c):
        perpendiculares.append("b y c")

    return {
        "producto_escalar": {
            "ab": ab,
            "ac": ac,
            "bc": bc
        },
        "producto_vectorial": {
            "axb": axb.tolist(),
            "bxc": bxc.tolist(),
            "axc": axc.tolist()
        },
        "paralelos": paralelos,
        "perpendiculares": perpendiculares
    }

# Vectores iniciales
a = [2, -1, 4]
b = [4, -2, 8]
c = [-5, 2, 1]

# Realizar las operaciones
resultados = operaciones_vectoriales(a, b, c)

# Imprimir los resultados
print("Producto escalar:")
print(f"a ¬∑ b = {resultados['producto_escalar']['ab']}")
print(f"a ¬∑ c = {resultados['producto_escalar']['ac']}")
print(f"b ¬∑ c = {resultados['producto_escalar']['bc']}")

print("\nProducto vectorial:")
print(f"a x b = {resultados['producto_vectorial']['axb']}")
print(f"b x c = {resultados['producto_vectorial']['bxc']}")
print(f"a x c = {resultados['producto_vectorial']['axc']}")

print("\nAn√°lisis:")
if resultados['paralelos']:
    print("Vectores paralelos:", ", ".join(resultados['paralelos']))
else:
    print("No hay vectores paralelos.")

if resultados['perpendiculares']:
    print("Vectores perpendiculares:", ", ".join(resultados['perpendiculares']))
else:
    print("No hay vectores perpendiculares.")
```
