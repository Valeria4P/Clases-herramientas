```python
# 1.- La secuencia del cortador de pizza perezoso (The Lazy Caterer's secuencia).
def lazy_caterer(n):
    f = [0] * (n + 1)
    f[0] = 1  # valor inicial

    for i in range(1, n + 1):
        f[i] = f[i - 1] + i

    return f

# Elegí 12 cortes porque es una cantidad suficientemente grande para ver cómo va aumentando el número de rebanadas,
#pero no tan grande como para que los resultados se vuelvan difíciles de entender.
#Con pocos cortes, como 1, 2 o 3, la secuencia se puede calcular fácilmente y no se aprecia mucho el crecimiento.
#En cambio, con 12 cortes se puede ver mejor el patrón sin que los números se vuelvan exagerados (aparte, y de manera personal, me gustan más los números pares)
n = 12
secuencia = lazy_caterer(n)

print(f"Secuencia del cortador de pizza perezoso hasta {n} cortes:")
for i, valor in enumerate(secuencia):
    print(f"f({i}) = {valor}")
```
