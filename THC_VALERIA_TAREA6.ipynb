```python
# 7. Calcular 2B + 3A, (B + 2A)**2, A**2 + 3B
a=np.array([[11,7,-1],[-3,5,2],[4,2,-2]])
b=np.array([[1,1,1],[3,10,-5],[-6,-10,-3]])
print("\nLa matriz A es:")
print(a)
print("\nLa matriz B es:")
print(b)

print("\n2B+3A es:")
ej_1=(b+b)+(a+a+a)
print(ej_1)

print("\n(B+2A)**2 es:")
ej_2=(a+a+b)@(a+a+b)
print(ej_2)

print("\nA**2+3B es:")
ej_3=(a@a)+(b+b+b)
print(ej_3)
```

```python
# 8.- Genere una funcion que reciba como entradas las dimensiones de una matriz T[n, m] (3 <= n, m <=5)
#Asegúrese de validar que sea cuadrada; de lo contrario, debe devolver el mensaje de error La matriz no
#es cuadrada y vuelva a preguntar las entradas. Posteriormente reemplace los valores con números aleatorios e imprima:

import numpy as np

def matriz_n_m():
    while True:
        try:
            n = int(input("Ingrese el numero de filas (3-5): "))
            m = int(input("Ingrese el numero de columnas (3-5): "))

            if not (3 <= n <= 5 and 3 <= m <= 5):
                print("Error: Las dimensiones deben estar entre 3 y 5.")
                continue

            if n != m:
                print("Error: La matriz no es cuadrada. Vuelva a intentar.")
                continue

            break
        except ValueError:
            print("Entrada inválida. Por favor, ingrese un número entero.")

    rango_min = 1
    rango_max = 10

    matriz = np.random.randint(rango_min, rango_max + 1, size=(n, m))
    print("\nLa matriz generada es:")
    print(matriz)

# La primera fila de manera inversa.
    print("\nLa primera fila de manera inversa es:")
    print(matriz[0, ::-1])

# La primera columna de manera inversa.
    print("\nLa primera columna de manera inversa es:")
    print(matriz[::-1, 0].reshape(-1, 1))

# El primer elemento y el  ́ultimo elemento.
    print(f"\nEl primer elemento es {matriz[0, 0]} y el último elemento es {matriz[n-1,m-1]}")

# El valor máximo, mínimo, el promedio de la diagonal y la traza.
    max_valor = np.max(matriz)
    min_valor = np.min(matriz)
    promedio_diagonal = np.mean(np.diag(matriz))
    traza = np.trace(matriz)
    print("\nEl valor máximo de la matriz es:", max_valor)
    print("El valor mínimo de la matriz es:", min_valor)
    print("El promedio de la diagonal de la matriz es:", promedio_diagonal)
    print("La traza de la matriz es:", traza)

# La suma de los valores de la primera fila y la última columna.
    suma_fila_columna = np.sum(matriz[0, :]) + np.sum(matriz[1:, -1])
    print("\nLa suma de los valores de la primera fila y la última columna es:", suma_fila_columna)

# La suma de los elementos pares recorriendo las filas.
    suma_pares_filas = 0
    for i in range(n):
        for j in range(m):
            if matriz[i, j] % 2 == 0:
                suma_pares_filas += matriz[i, j]
    print("\nLa suma de los elementos pares recorriendo las filas es:", suma_pares_filas)

# La suma de los elementos impares recorriendo las columas.
    suma_impares_columnas = 0
    for j in range(m):
        for i in range(n):
            if matriz[i, j] % 2 != 0:
                suma_impares_columnas += matriz[i, j]
    print("La suma de los elementos impares recorriendo las columnas es:", suma_impares_columnas)

# Imprima los promedios de las filas y columnas.
    print("\nPromedios de las filas:")
    for i in range(n):
        print(f"Fila {i+1}: {np.mean(matriz[i, :])}")

    print("\nPromedios de las columnas:")
    for j in range(m):
        print(f"Columna {j+1}: {np.mean(matriz[:, j])}")

# La antidiagonal y el promedio de sus valores.
    antidiagonal = np.diag(np.fliplr(matriz))
    promedio_antidiagonal = np.mean(antidiagonal)
    print("\nLa antidiagonal es:", antidiagonal)
    print("El promedio de la antidiagonal es:", promedio_antidiagonal)
matriz_n_m()
```

```python
# 1. En la clase de ecuaciones diferenciales se pide a los estudiantes encontrar la solución del problema de
# valor inicial: y = −y + x y(−1) = 5

# Los estudiantes llegaron a la solución general cuya ecuación es:
# y = (x − 1) + Ce−x


import numpy as np
import matplotlib.pyplot as plt

# Define la función de la solución general
def y(x, C):
    return (x - 1) + C * np.exp(-x)

# Define el rango de valores x
x = np.linspace(-5, 5, 400)

# Grafica la curva para C > 0
plt.plot(x, y(x, 1), label='C > 0', color='blue')

# Grafica la curva para C = 0
plt.plot(x, y(x, 0), label='C = 0', color='black', linewidth=2)

# Grafica la curva para C < 0
plt.plot(x, y(x, -1), label='C < 0', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Familia de curvas para la solución de la ecuación diferencial')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Sea la función: f(x) = (3x + 7) / (x^2 - 4)

# Utilizando la librería matplotlib:

import numpy as np
import matplotlib.pyplot as plt

# Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras.
def f(x):
    return ((3*x) + 7) / ((x**2) - 4)

x = np.linspace(-10, 10, 2000)
y = f(x)
plt.figure(figsize=(10, 6))

plt.plot(x, y, label='f(x) = (3x + 7) / (x^2 - 4)')


# Grafique las asíntotas horizontales y verticales.
plt.axvline(x=-2, color='r', linestyle='--', label='Asíntota vertical x = -2')
plt.axvline(x=2, color='g', linestyle='--', label='Asíntota vertical x = 2')

plt.axhline(y=0, color='y', linestyle='--', label='Asíntota horizontal y = 0')


# Defina los límites del eje y para evitar que la gráfica muestre valores muy altos o muy bajos cerca de las asíntotas verticales.
plt.ylim(-5, 5)

# Añada una leyenda indicando los elementos presentes en la gráfica.
plt.legend()

plt.xlabel('x')
plt.ylabel('y')

plt.title('Gráfica de f(x) = $(3x + 7) / (x^2 - 4)$')

# Muestra la gráfica
plt.grid(True)
plt.show()
```
